```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
age = np.sort(np.random.randint(10,25,5))
print(age)
print(type(age))
```

```python
weight = np.sort(np.random.randint(25,40,5))
print(weight)
```

```python
plt.scatter(x=age,y=weight)
plt.xlabel('Age')
plt.ylabel('Weight')
plt.title('scatter plot with age and weight')
plt.show()
```

```python
plt.plot(age,weight,marker='o',c='green',ms=20)
plt.xlabel('Age')
plt.ylabel('Weight')
plt.title('scatter plot with age and weight')
plt.show()
```

```python
x = np.radians(np.arange(0,1441,15))
s = np.sin(x)
c = np.cos(x)
plt.plot(x,s,ls='--')
plt.plot(x,c,ls='--')
plt.show()
```
