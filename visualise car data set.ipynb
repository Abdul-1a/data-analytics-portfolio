```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
plt.scatter(x=cars.price,y=cars.mileage,s=20)
plt.xlabel('Price')
plt.ylabel('Mileage')
plt.show()
```

```python
# another method to assign data to variables
plt.scatter(x='price',y='mileage',data=cars,s=20)
plt.xlabel('Price')
plt.ylabel('Mileage')
plt.title('scatter plot for cars data')
plt.show()
```

```python
plt.scatter(x='price',y='mileage',data=cars[cars.model=='SES'],label='SES',marker="*")
plt.scatter(x='price',y='mileage',data=cars[cars.model=='SE'],label='SE',marker="^")
plt.scatter(x='price',y='mileage',data=cars[cars.model=='SEL'],label='SEL',marker='d')
plt.legend(loc='best')
plt.show()
```

```python
models = {'SES':'blue','SE':'yellow','SEL':'red'}
for m in models:
    plt.scatter(x='price',y='mileage',data=cars[cars.model==m],label=m,c=models[m])
plt.legend()
plt.show()
```

```python
models = set(cars.transmission)
for m in models:
    plt.scatter(x='price',y='mileage',data=cars[cars.transmission==m],label=m)
plt.legend()
plt.show()
```
